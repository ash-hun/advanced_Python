# Bow

```python
sentence = "There are many variations of passages of Lorem Ipsum available, but the majority have suffered alteration in some form, by injected humour, or randomised words which don't look even slightly believable. If you are going to use a passage of Lorem Ipsum, you need to be sure there isn't anything embarrassing hidden in the middle of text. All the Lorem Ipsum generators on the Internet tend to repeat predefined chunks as necessary, making this the first true generator on the Internet. It uses a dictionary of over 200 Latin words, combined with a handful of model sentence structures, to generate Lorem Ipsum which looks reasonable. The generated Lorem Ipsum is therefore always free from repetition, injected humour, or non-characteristic words etc."
```

```python
# BoW를 만드는 방법은 다음과 같다.
# 1. 문자열을 띄어쓰기를 기준으로 쪼갠다.
# 2. 문장 안의 각 단어에 고유한 정수 인덱스를 부여한다.
# 3. 각 인덱스의 위치에 단어 토큰의 등장 횟수를 기록한 벡터를 만든다.
```

```python
# 1. 문자열을 띄어쓰기를 기준으로 쪼갠다.
temp = list(sentence.split(" "))
temp
```

```python
#2. 문장 안의 각 단어에 고유한 정수 인덱스를 부여한다.
cnt = {}
for idx, val in enumerate(temp):
    cnt[str(val)] = int(idx)
cnt
```

```python
# 3. 각 인덱스의 위치에 단어 토큰의 등장 횟수를 기록한 벡터를 만든다.
sentence_bow = []
for i, v in cnt.items():
    sentence_bow.append(v)
sentence_bow
```

## BoW 패키지 : CountVectorizer
- ScikitLearn에서 제공하는 클래스로, 단어의 빈도를 Count하여 Vector로 만드는 CountVectorizer 클래스로 영어에 한해서 손쉽게 BoW를 만들 수 있음.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentence2 = [sentence]
sentence2
```

```python
vector = CountVectorizer()
vector.fit_transform(sentence2).toarray()
```
